```python

import matplotlib.pyplot as plt
import numpy as np

def quadratic_solution(a, b, c, x):
    return a * x**2 + b * x + c

def read_coefficients_from_file(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        a, b, c = map(float, lines)
    return a, b, c
filename = 'wv.csv'
a, b, c = read_coefficients_from_file(filename)

x_values = np.linspace(-10, 10, 100)

# Calculate corresponding y values using the quadratic solution
y_values = quadratic_solution(a, b, c, x_values)

plt.plot(x_values, y_values, label=f'{a}x^2 + {b}x + {c}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Solution (Read from File)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
def quadratic_solver(a, b, c):
    delta = b**2 - 4*a*c
    if delta >= 0:
        x1 = (-b + (delta**0.5)) / (2*a)
        x2 = (-b - (delta**0.5)) / (2*a)
        return x1, x2
    else:
        return None
def plot_quadratic_equations(equations):
    x = np.linspace(-10, 10, 100)
    for equation in equations:
        a, b, c, solutions = equation
        y = a*x**2 + b*x + c
        plt.plot(x, y, label=f'{a}x^2 + {b}x + {c}')
        if solutions:
            plt.scatter(solutions, [0, 0], color='red', marker='x')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.title('Quadratic Equations for temperature and Solutions')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()
equations = []
while True:
    try:
        a_input = input("Enter coefficient a (or type 'exit' to stop): ")
        if a_input.lower() == 'exit':
            print("Exiting the program.")
            break
        a = float(a_input)
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))
        solutions = quadratic_solver(a, b, c)
        print("Solutions:", solutions)
        equations.append((a, b, c, solutions))
    except ValueError:
        print("Invalid input. Please enter valid numeric coefficients.")
plot_quadratic_equations(equations)

2
```

```python
def quadratic_solution(a, b, c, x):
    y = a * x**2 + b * x + c
    return y

def plot_temperature_over_time(a, b, c, time_range):
    time = np.array(time_range)
    temperature = quadratic_solution(a, b, c, time)

    plt.plot(time, temperature, label=f'a={a}, b={b}, c={c}')

def main():
    print("Weather Modeling using Quadratic Solution")


    num_sets = int(input("Enter the number of sets: "))

    for i in range(num_sets):

        if i == 0:
            a = float(input("Enter the value of a: "))
            b = float(input("Enter the value of b: "))
            c = float(input("Enter the value of c: "))
        else:
            a = 10
            b = 5
            c = 70


        time_range = np.linspace(0, 10, 100)


        plot_temperature_over_time(a, b, c, time_range)
    plt.legend()
    plt.title("Temperature Over Time for Multiple Sets")
    plt.xlabel("Time")
    plt.ylabel("Temperature")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
